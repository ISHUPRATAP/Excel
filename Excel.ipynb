QUESTION 1

Paste:
Pastes everything exactly as copied.
Example: If you copy a cell with a formula, formatting, and value, Excel pastes all of it. No questions asked.

Paste Special:
Lets you choose what to paste like values, formulas, formats, or transpose data.
Example: Copy a formula result and use Paste Special → Values to paste only the number, not the formula.

Short version:
Paste = everything.
Paste Special = only what you want.

QUESTION 2

1. Freeze Panes vs Split Panes in Excel

Freeze Panes: Locks selected rows/columns so they remain visible while scrolling.
Usefulness: Helpful for keeping headers visible in large worksheets.

Split Panes: Divides the worksheet into separate scrollable sections.
Usefulness: Useful for comparing different parts of the same worksheet at once.

QUESTION 3

Inserting a New Row vs Inserting a New Column in Excel

Insert Row: Adds a horizontal line of cells above the selected row.
Example: Inserting a row above Row 5 shifts Row 5 and below downward.

Insert Column: Adds a vertical line of cells to the left of the selected column.
Example: Inserting a column before Column C shifts Column C and to the right.

Multiple Rows/Columns: Yes, you can insert multiple rows or columns at once by selecting multiple rows or columns first, then choosing Insert.

QUESTION 4

Logical functions in Excel are functions that evaluate conditions (TRUE or FALSE) and return results based on those conditions.

Examples:

IF()

Use: Returns one value if a condition is true, another if false.

Example: =IF(A1>=40,"Pass","Fail") → Checks if marks are 40 or more.

AND()

Use: Returns TRUE if all conditions are true.

Example: =AND(A1>=40, B1>=40) → Checks if a student passed in both subjects.

QUESTION 5

Purpose of XLOOKUP:
XLOOKUP is used to search for a value in a range and return a corresponding value from another range. It is more flexible and powerful than VLOOKUP.

Difference between XLOOKUP and VLOOKUP:

Search direction: XLOOKUP can search left, right, up, or down, while VLOOKUP can only search left to right.

Column index: XLOOKUP does not require a column number; VLOOKUP does.

Error handling: XLOOKUP has built-in error handling (e.g., return “Not Found”).

Exact match: XLOOKUP uses exact match by default, while VLOOKUP needs an option.

Example:

XLOOKUP(A2, A:A, B:B) → finds value in column A and returns matching value from column B.
